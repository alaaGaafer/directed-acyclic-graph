```python
%matplotlib inline 
import networkx as nx
import matplotlib.pyplot as plt
plt_1 = plt.figure(figsize=(11, 11))
G = nx.DiGraph()
#adding nodes to the graph
G.add_node("Manager")
G.add_node("Depratment X head")
G.add_node("Depratment Y head")
G.add_node("Staff A")
G.add_node("Staff B")
G.add_node("Staff C")
G.add_node("Staff D")
G.add_node("Staff E")
#connecting the nodes by edeges
G.add_edge("Manager", "Depratment X head")
G.add_edge("Manager", "Depratment Y head")
G.add_edge("Depratment X head","Staff A")
G.add_edge("Depratment X head","Staff B")
G.add_edge("Depratment Y head","Staff C")
G.add_edge("Depratment Y head","Staff D")
G.add_edge("Depratment Y head","Staff E")
#assigning postitions to the nodes
pos = {"Manager":(11, 11),
 "Depratment X head":(7.5, 7.5), "Depratment Y head":(14, 7.5),
 "Staff A":(6, 6), "Staff B":(9, 6),
 "Staff C":(12, 6), "Staff D":(15, 6), "Staff E":(18, 6)}
colors = ["skyblue", "green","green", "orange","orange","orange","orange","orange"]
sizes = [10000, 15000, 15000, 10000, 10000, 10000, 10000, 10000]

nx.draw_networkx(G, pos = pos,with_labels = True, arrows = True,
                 node_shape = "o", node_size = sizes,
                 node_color = colors,
                 edgecolors = "gray")
plt.show()
```

```python
print("Number of nodes=",G.number_of_nodes())
print("Number of edges=",G.number_of_edges())
start_nodes = [n for n, d in G.in_degree() if d == 0]
end_nodes = [n for n, d in G.out_degree() if d == 0]
print("Start node:", start_nodes)
print("End nodes:", end_nodes)
```

```python
path1 = nx.all_simple_paths(G, source="Manager", target="Staff A")
print(list(path1))
path1 = nx.all_simple_paths(G, source="Manager", target="Staff B")
print(list(path1))
path1 = nx.all_simple_paths(G, source="Manager", target="Staff C")
print(list(path1))
path1 = nx.all_simple_paths(G, source="Manager", target="Staff D")
print(list(path1))
path1 = nx.all_simple_paths(G, source="Manager", target="Staff E")
print(list(path1))
```
